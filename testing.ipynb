```python
from fMRIsim import fMRIsim
import matplotlib.pyplot as plt

sim = fMRIsim()
sim.nevents = 5
sim.length = 'medium'
sim.has_integrator = True
sim.te = [0.15, 0.35, 0.5]
sim.snr = 8
sim.simulate()
```

```python
plt.figure(num=1, figsize=(16, 9), dpi=300)
plt.plot(sim.innovation)
plt.plot(sim.r2)
plt.plot(-sim.bold)
plt.show()
```
